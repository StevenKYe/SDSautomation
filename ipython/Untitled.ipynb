```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('dark_background')
%matplotlib inline
```

```python
NUMFRAMES = 200
fig = plt.figure(facecolor='black', figsize=(6,4))
ax = fig.add_subplot(111)
line, = ax.plot([], [], 'r-')

def animate(i):
    # Update the start/end point
    x = np.linspace(0, NUMFRAMES/5, 1000)
    y = np.sin(2 * np.pi * (x + 0.1 * i))
    line.set_data(x, y)
    return line,

ani = FuncAnimation(fig, animate, frames=NUMFRAMES, interval=100, blit=True)
```

```python
fig = plt.figure()
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))
line, = ax.plot([], [], '#00B0F0', lw=3)
ax.axis('off')

def init():
    line.set_data([], [])
    return line,
def animate(i):
    x = np.linspace(0, 4, 1000)
    y = np.sin(2 * np.pi * (x - 0.03 * i))
    line.set_data(x, y)
    return line,

anim = FuncAnimation(fig, animate, frames=200, interval=20, blit=True)


anim.save('sine_wave_blue.gif', writer='imagemagick', savefig_kwargs={'facecolor': 'black'})
```
